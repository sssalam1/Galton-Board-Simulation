# Galton Board Simulation

```python
# Importing the libraries
import matplotlib.pyplot as plt
import random
```

```python
# Giving the input of number of balls and the bins
balls = 100
lines = 15
```

```python
# Creating the empty bins
dist = [0] * (lines*2)
for ball in range(balls):
    total = 0
    for line in range(lines):
        outcome = random.uniform(0,1)
        if outcome >= 0.5:
            total = total + 1
        else:
            total = total - 1
    dist[total + lines ] = dist[total + lines ] + 1
   
for i in range(lines):
    dist[i] = dist[i*2]
```

```python
plt.bar([str(i) for i in range(lines)], dist[0:lines])
plt.xticks([])
plt.show()
```

```python
plt.plot([str(i) for i in range(lines)], dist[0:lines],marker='o',linestyle='None')
plt.xticks([])
plt.show()

(markers, stemlines, baseline) = plt.stem([str(i) for i in range(lines)], dist[0:lines],use_line_collection=True)
```
